# 20. Line Chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create data points for the x axis
# arguments: starting point, ending point, and how many points in between
x = np.linspace(0,10,10)
# next, let's create a sin wave
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Some function of time")
plt.title("My cool chart")
```

```python
# same example with more points
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Some function of time")
plt.title("My cool chart")
```

# 21. Scatterplot

What we are really trying to do in machine learning is function approximation.

https://github.com/lazyprogrammer/machine_learning_examples/tree/master/linear_regression_class

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the csv and turn it into a matrix
A = pd.read_csv('data_1d.csv', header=None).values
```

```python
x = A[:,0]
y = A[:,1]
```

```python
# now we're ready to do a scatterplot
plt.scatter(x, y)
plt.show()
```

We happen to know what the exact function is that fits the scatterplot, so let's add it to che chart.

```python
x_line = np.linspace(0,100,100)
y_line = 2*x_line + 1
```

```python
plt.scatter(x, y)
plt.plot(x_line, y_line)
plt.show()
```

# 22. Histogram

```python
plt.hist(x)
plt.show()
```

```python
R = np.random.random(10000)
plt.hist(R)
plt.show()
```

Control the number of buckets

```python
plt.hist(R, bins=20)
plt.show()
```

One assumption that we make when we're doing linear regression is that the error is normally distributed.

So if we take all the y data points and subtract it from its corresponding place in the fitted line, then that data should be normally distributed.

We can check that with a histogram.

```python
y_actual = 2*x + 1
residuals = y - y_actual
plt.hist(residuals)
plt.show()
```
